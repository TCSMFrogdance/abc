```python
import sqlite3
import pandas as pd
```

```python
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
    except Error as e:
        print(e)
    return conn
```

```python
def select_all(conn):
    """
    Query all rows in the database
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    cur.execute("""SELECT * FROM Sales
              JOIN Products ON Sales.ProductID = Products.ProductID
              JOIN Stores ON Sales.StoreID = Stores.StoreID
              JOIN Employees ON Sales.SalesPersonID = Employees.SalesPersonID
              JOIN Customers ON Sales.CustomerID = Customers.CustomerID""")
    rows = cur.fetchall()
    return rows
```

```python
def select_tableName(conn, tableName):
    """
    Query tableName
    :param conn: the Connection object
    :param tableName:
    :return: string
    """
    cur = conn.cursor()
    cur.execute("SELECT * FROM {}".format(tableName))
    rows = cur.fetchall()
    return rows
```

```python
def select_tableName_join_tableName(conn, tableName1, tableName2, priority):
    """
    Querry sales with priority
    :param conn: the connection object
    :param tableName1: string
    :param tableName2: string
    :param priority: string
    
    :return:
    """
    cur = conn.cursor()
    cur.execute("SELECT * FROM {} JOIN {} ON Sales.{} = {}.{}".format(tableName1, tableName2, tableName1, priority, tableName2, priority))
    rows = cur.fetchall()
    return rows
```

```python
def select_sum_sale_by_priority(db, tableName, priority, key):
    """
    Qeurry sum sales with priority
    :param conn: the connection
    :param priority: string
    :param key: string
    """
    cur = db.cursor()
    cur.execute("""SELECT {}.{}, SUM(Sales.Sales) FROM Sales
              JOIN  {} ON Sales.{} = {}.{}
              GROUP BY {}.{}""".format(tableName, priority, tableName, key, tableName, key, tableName, priority))
    rows = cur.fetchall()
    return rows
```
